```python
import torch
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
lr = 0.02

w = torch.Tensor([1.0])
w.requires_grad = True  # 表示w是需要计算梯度的，默认对变量不计算梯度
```

```python
def forward(x):
    return x * w  # 注意这里的*, 因为w是一个tensor的数据类型，所以这里的*是一个tenosr类型之间的数乘，所以会自动对x转换成tensor数据类型
```

```python
def loss(x, y):
    y_pre = forward(x)
    return (y_pre - y) ** 2
```

```python
print("predict(before training)", forward(4).item())

for epoch in range(150):
    for x, y in zip(x_data, y_data):
        l = loss(x, y)  # 前馈只需要使用SGD计算loss

        l.backward()  # 只要执行了backward计算图就被释放了，下次调用再次构建，因为在复杂的网络中每次调用的计算图可能不一样
        print("\tgrant:", x, y, w.grad.item())
        w.data = w.data - lr * w.grad.data
        w.grad.data.zero_()  # 每次更新完后梯度需要清零

    print("progress:", "epoch:", epoch, "loss:", l.item())
print("predict(after training)", 4, forward(4).item())
```
